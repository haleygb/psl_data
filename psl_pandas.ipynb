```python
import pandas as pd
from tabulate import tabulate

#read in the CSV file
psl_data = pd.read_csv(r"/Users/haleybohn/psl.csv")

#check the first few rows of the dataframe
print(tabulate(psl_data.head(), headers='keys', tablefmt='psql'))
```

```python

syll = [1, 2, 3]

#Show rows with nasal vowels in words with 1, 2, or 3 syllables
psl_nasal = psl_data[(psl_data['word'].str.contains('am|em|im|om|um|un|on|in|an') | psl_data['pro'].str.contains('~')) & psl_data['nSyl'].isin(syll)]

#Show the first 100 rows of the filtered dataframe
print(tabulate(psl_nasal.head(100), headers='keys', tablefmt='psql'))
```

```python
#Specify the columns to include in the new CSV file
words_with_nasal_vowels = psl_nasal[['word', 'pro', 'nSyl', 'nSylPre', 'weightProfile', 'wordProfile', 'finalWordProfile']]

#Export the filtered dataframe to a new CSV file
words_with_nasal_vowels.to_csv(r"/Users/haleybohn/psl_nasal.csv", index=False)
```
